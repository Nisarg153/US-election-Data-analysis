# Checking the NaN values

```python
election_df.isna().sum()
```

As we have total records 4287 in the dataset and in the notes column there are also 4287 NaN values.
it indicates that we do not need notes column. 
Office and version has the same value all along. 
so we can drop notes, office and version columns.

```python
election_df.drop(['notes', 'office', 'version'], axis = 1, inplace=True)
```

```python
election_df.isna().sum()
```
